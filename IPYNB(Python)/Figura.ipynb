```python
import matplotlib as mtl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection = '3d')

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

x = 0.5 * np.outer(np.cos(u), np.sin(v))+ 0
y = 0.5 * np.outer(np.sin(u), np.sin(v))+ 0
z = 0.5 * np.outer(np.ones(np.size(u)), np.cos(v))+ 0

x1 = 0.5 * np.outer(np.cos(u), np.sin(v))+ 0.5
y1 = 0.5 * np.outer(np.sin(u), np.sin(v))+ 0.5
z1 = 0.5 * np.outer(np.ones(np.size(u)), np.cos(v))+ 0

x2 = 0.5 * np.outer(np.cos(u), np.sin(v))+ 0
y2 = 0.5 * np.outer(np.sin(u), np.sin(v))+ 0.5
z2 = 0.5 * np.outer(np.ones(np.size(u)), np.cos(v))+ 0.5
ax.plot_surface(x2, y2, z2, rstride=3, cstride=3, alpha=0.7)
ax.plot_surface(x, y, z, rstride=3, cstride=3, alpha=0.7)
ax.plot_surface(x1, y1, z1, rstride=3, cstride=3, alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# prepare some coordinates
x, y, z = np.indices((5, 5, 8))
# draw cuboids in the top left and bottom right corners, and a link between them
cube1 = (x < 5) & (y < 5) & (z < 5)
cube2 = (x < 5)& (x >= 3) & (y <= 2)& (y >= 0) & (z <= 3)
cube3 = (x < 3) & (x > 0)& (y < 3) &(y > 0)& (z >= 3)

# combine the objects into a single boolean array
voxels = cube1 | cube2 | cube3

# set the colors of each object
colors = np.empty(voxels.shape, dtype=object)
colors[cube3] = 'red'
colors[cube1] = 'yellow'
colors[cube2] = 'blue'

# and plot everything
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.voxels(voxels, facecolors=colors, linewidths=1,edgecolors='grey',alpha=.25)

plt.show()
```
