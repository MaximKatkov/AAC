```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

p,q=np.mgrid[0:2*np.pi:150j,0:2*np.pi:150j ]
x=np.cos(p)*np.sin(q)
y=np.sin(p)*np.sin(q)
z=np.cos(q)
fig=plt.figure()
ax=fig.add_subplot( projection='3d')
ax.plot_surface(x,y,z,color='b',alpha=0.2)
ax.plot_surface(x-1,y,z,color='b', alpha=0.2)
ax.plot_surface(x-0.5,y,z-1,color='b', alpha=0.2)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

x, y, z = np.indices((5, 5, 8))
cube1 = (x < 5) & (y < 5) & (z < 5)
cube2 = (x < 5)& (x >= 3) & (y <= 2)& (y >= 0) & (z <= 3)
cube3 = (x < 3) & (x > 0)& (y < 3) &(y > 0)& (z >= 3)

# combine the objects into a single boolean array
voxels = cube1 | cube2 | cube3

colors = np.empty(voxels.shape, dtype=object)
colors[cube3] = 'green'
colors[cube1] = 'red'
colors[cube2] = 'blue'

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.voxels(voxels, facecolors=colors,edgecolors='grey',alpha=.25)

plt.show()
```
