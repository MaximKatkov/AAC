```python
import numpy as np

def lenght(x0):
    length_x0 = 0
    while(x0 > 0):
        length_x0 +=1
        x0 = x0 // 10
    return length_x0

def generator(x0):
    y0=x0**(2)
    length_x0=lenght(x0)
    length_y0=lenght(y0)
# Колличество отступов в разрядах до середины
    n=(length_y0-length_x0)//2
    if n%2==0:
    #отступ справа чтобы сократить число справа на n порядков
        right=10**n
    #отступ слева чтобы при делении на 10^N осталось значение числа с длиной до 10^length_y0-n
        left=10**(length_y0-n)
        y0=y0%left
        y0=y0//right
    else:
        right=10**(n+1)
        left=10**(length_y0-n)
        y0=y0%left
       
        y0=y0//right
    return y0

def GENERATOR(x0,n,a,b,c):
    R_Number=[]
    M=2**(25)
    for i in range(n):
        r0=generator(x0)
        r=(a*r0**(2)+b*r0+c)%M
        R_Number.append(r)
        x0=r
    print('Список рандомных чисел:\n',R_Number,'\n')
    stop=0
    for i in range(n):
        for j in range(n):
            if R_Number[i]==R_Number[j] and i!=j:
                print(' Период T=',abs(i-j),'\n','Позиция с которой начинается период: ',i,'\n','Значение с которого начинается зацикливание: ',R_Number[i])
                stop=i-j
            if stop!=0:
                break
    return R_Number
    
print('\t','МЕТОД СЕРЕДИННЫХ КВАДРАТОВ\n')
x0=int(input('Введите начальное значение X0='))
N=int(input('Введите колличество итераций n='))
a=675
b=456
c=345
print('\n')
R_Number=GENERATOR(x0,N,a,b,c)

#************************* Методом Слияния ***********************************#

def MERGE(arr):
    # будем сортировать на два подмассива ,пока размер подмножества не будет меньше 2
    if len(arr) > 1:
        mid = len(arr) // 2
        # от 0 до середины
        left = arr[:mid]
        # от середины до конца
        right = arr[mid:]
        # С помощбю рекурсии будем сортировать каждую половину пока работает условие выше
        MERGE(left)
        MERGE(right)

        i =0
        j =0
        k =0
        while i <len(left) and j <len(right):
            # сравниваем элементы в каждой части и добавляем их в список
            if left[i]<right[j]:
                arr[k]=left[i]
                i+=1
            else:
                arr[k]=right[j]
                j+=1
            k+=1
        # обЪелиняем отсортированные части
        while i <len(left):
            arr[k] = left[i]
            i+=1
            k+=1
            
        while j <len(right):
            arr[k]= right[j]
            j+=1
            k+=1

    return arr
 
Sort_list= MERGE(R_Number)
print('МЕТОД СЛИЯНИЯ','\n',' Отстортированный массив:','\n',Sort_list,'\n')

#******************** Методом Подсчёта *******************#

List=[1,7,12,4,8,0,7,73,3,6,24,5,21,11,8,2]
def COUNT(arr):
    # находим максимальное значение в массиве
    max_var = arr[0]
    for var in arr:
        if var>max_var:
            max_var=var
    # создаем массив для подсчета количества вхождений
    counts = np.zeros(max_var + 1,dtype=int) 
    for var in arr:
        # увеличиваем счетчик для соответствующего значения
        counts[var] += 1

    sorted_arr = []
    for i in range(max_var + 1):
        part_arr=[]
        for j in range(counts[i]):
            part_arr.append(i) 
        # добавляем соответствующее количество элементов в отсортированный массив
        sorted_arr+=part_arr 

    return sorted_arr

Sort= COUNT(List)
print('МЕТОД ПОДСЧЁТА','\n',' Отстортированный массив:','\n',Sort)
```
