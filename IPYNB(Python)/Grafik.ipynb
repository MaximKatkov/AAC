```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

p,q=np.mgrid[0:2*np.pi:150j,0:2*np.pi:150j ]
x=np.cos(p)*np.sin(q)
y=np.sin(p)*np.sin(q)
z=np.cos(q)
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot( projection='3d')
ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.set_zlim(-2,2)
ax.plot_surface(x,y,z,color='b',alpha=0.2)
ax.plot_surface(x-1,y,z,color='b', alpha=0.2)
ax.plot_surface(x-0.5,y,z-1,color='b', alpha=0.2)
```
