```python
import itertools

import networkx as nx
import numpy.random as rnd
import matplotlib.pyplot as plt
```

```python
graph = nx.Graph()

graph
```

```python
graph.add_node('A')
graph.add_node('B')
graph.add_node('C')

graph.nodes()
```

```python
def add_edge(f_item, s_item, graph=None):
  graph.add_edge(f_item, s_item)
  graph.add_edge(s_item, f_item) 
```

```python
add_edge('A', 'B', graph=graph)
add_edge('B', 'C', graph=graph)
add_edge('B', 'D', graph=graph)
add_edge('D', 'E', graph=graph)
nc=[3,1,4,7,2]
ns=[1000,1500, 700, 1000,1500]
nx.draw_circular(graph,
         node_color=nc,
         node_size=ns,
         with_labels=True)
```

```python
cities = {'A':(0, 20),
     'B':(15, 24),
     'C':(16, 41),
     'D':(10, 40)}

graph = nx.Graph()
graph.add_nodes_from(cities)
```

```python
kilometres = {('A', 'B', 15),
              ('B', 'C', 16),
              ('B', 'D', 25),
              ('C', 'D', 14),
              ('D', 'A', 18)}

graph.add_weighted_edges_from(kilometres)
```

```python
#import Matplotlib либо GraphDot.
nx.draw_circular(graph,
         node_color='red',
         node_size=1000,
         with_labels=True)
```

```python
def complete_graph(N: int) -> nx.Graph:
  graph = nx.Graph()
   
  N_range = range(N)
   
  all_pairs = itertools.permutations(N_range, 2)
   
  graph.add_nodes_from(N_range)
  graph.add_edges_from(all_pairs)
   
  return graph
```

```python
graph = complete_graph(15)

nx.draw_circular(graph, 
         node_color='y',
         node_size=750,
         with_labels=True)
```

```python
import telebot;
bot = telebot.TeleBot('%ваш токен%');
```

```python
@bot.message_handler(content_types=['text'])# @bot.message_handler(content_types=['text', 'document', 'audio'])
def get_text_messages(message):
    if message.text == "Привет":
        bot.send_message(message.from_user.id, "Привет, чем я могу тебе помочь?")
    elif message.text == "/help":
        bot.send_message(message.from_user.id, "Напиши привет")
    else:
        bot.send_message(message.from_user.id, "Я тебя не понимаю. Напиши /help.")
```
